```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
from sklearn.datasets import make_circles
```

```python
X, y = make_circles(n_samples=1000, noise=0.05, random_state=47)

X[:,0:]

plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.bwr)
```

```python
X,y
```

```python
plt.scatter(X[:, 0], y, c=y, cmap=plt.cm.bwr)
```

```python
from sklearn.impute import SimpleImputer

impute = SimpleImputer(strategy="mean")
impute.fit(X[:, :])
X[:, :] = impute.transform(X[:, :])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

scaler = StandardScaler()
X_train_scal = scaler.fit_transform(X_train)
X_test_scal = scaler.transform(X_test)

X_test,X_train,X_test_scal,X_train_scal
```

```python
from sklearn.linear_model import LogisticRegression
log = LogisticRegression()

log.fit(X_train_scal,y_train)
y_pred = log.predict(X_test_scal)

from sklearn.metrics import accuracy_score , confusion_matrix

acc = accuracy_score(y_test,y_pred)
cm = confusion_matrix(y_test,y_pred)
print(f'Logistic regression :\n Accuracy score:{acc},Confussion matrix:\n{cm}')
sns.heatmap(cm, annot = True, cbar = True)
```

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)
y_pred = dt.predict(X_test)


acc = accuracy_score(y_test,y_pred)
cm = confusion_matrix(y_test,y_pred)
print(f'Decision Tree :\n Accuracy score:{acc},\nConfussion matrix:\n{cm}')
sns.heatmap(cm, fmt='d',annot = True, cbar = True)


```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=100)
rf.fit(X_train, y_train)
y_pred = rf.predict(X_test)
acc = accuracy_score(y_test,y_pred)
cm = confusion_matrix(y_test,y_pred)
print(f'Random Forrest:\n Accuracy score:{acc},\nConfussion matrix:\n{cm}')
sns.heatmap(cm, annot = True, cbar = True,fmt='d',)
```

```python
from sklearn.naive_bayes import GaussianNB

nb = GaussianNB()
nb.fit(X_train_scal, y_train)
y_pred = nb.predict(X_test_scal)
acc = accuracy_score(y_test,y_pred)
cm = confusion_matrix(y_test,y_pred)
print(f'Naive bayes :\n Accuracy score:{acc},\nConfussion matrix:\n{cm}')
sns.heatmap(cm, annot = True, cbar = True,fmt='d',)

```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train_scal, y_train)
y_pred = knn.predict(X_test_scal)
acc = accuracy_score(y_test,y_pred)
cm = confusion_matrix(y_test,y_pred)
print(f'KNN :\n Accuracy score:{acc},\nConfussion matrix:\n{cm}')
sns.heatmap(cm, annot = True, cbar = True,fmt='d',)
```
