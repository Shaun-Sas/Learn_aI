```python
import pandas as pd

df = pd.read_csv('data.csv')
print(df.head(10))

# Remove rows with NaN values
df = df.dropna()
print(df.head(10))
```

```python
df.info()
```

```python
inputs = df.iloc[:, 0:2].values
inputs.T
```

```python

plt.scatter(inputs[:, 0], inputs[:, 1])
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = sc.fit_transform(inputs)
X.T
```

```python
import scipy.cluster.hierarchy as sch

plt.figure(figsize = (20, 10))

z = sch.linkage(X, method = 'ward')
dendrogram = sch.dendrogram(z)
plt.title('Dendrogram')
```

```python
from sklearn.cluster import AgglomerativeClustering

hc = AgglomerativeClustering(n_clusters = 3, metric = 'euclidean', linkage = 'ward')
y_hc = hc.fit_predict(X)
```

```python
y_hc
```

```python
import numpy as np
import matplotlib.pyplot as plt

n_clusters = len(np.unique(y_hc))

centroids = np.zeros((n_clusters, X.shape[1]))
for i in range(n_clusters):
    centroids[i] = X[y_hc == i].mean(axis=0)

plt.figure(figsize=(10, 6))
plt.scatter(X[:, 0], X[:, 1], c=y_hc, cmap='viridis', s=50, label='Data Points')
plt.scatter(centroids[:, 0], centroids[:, 1], s=200, c='red', marker='X', label='Centroids')
plt.colorbar(label='Cluster Label')
plt.title('Scatter Plot of Clusters with Centroids')
plt.xlabel('Culmen length')
plt.ylabel('Culmen depth')
plt.legend()
plt.show()
```

```python
# Divisive clustering

import numpy as np
from sklearn.cluster import KMeans

def divisive_clustering(X, n_clusters):
    # Start with all points in a single cluster
    labels = np.zeros(X.shape[0], dtype=int)
    cluster_list = [(X, np.arange(X.shape[0]))]
    current_cluster_id = 0

    while len(cluster_list) < n_clusters:
        # Find the largest cluster to split
        largest_cluster_points, largest_cluster_indices = max(cluster_list, key=lambda item: len(item[0]))

        # Perform k-means clustering with 2 clusters
        kmeans = KMeans(n_clusters=3, random_state=0)
        sub_labels = kmeans.fit_predict(largest_cluster_points)

        # Remove the largest cluster from the list
        cluster_list.remove((largest_cluster_points, largest_cluster_indices))

        # Create two new clusters from the split
        cluster_list.append((largest_cluster_points[sub_labels == 0], largest_cluster_indices[sub_labels == 0]))
        cluster_list.append((largest_cluster_points[sub_labels == 1], largest_cluster_indices[sub_labels == 1]))

        # Re-assign the labels in the overall labels array
        for i, (points, indices) in enumerate(cluster_list):
            labels[indices] = i

        # Update current cluster ID to match the number of clusters
        current_cluster_id = len(cluster_list) - 1

    return labels
```

```python
y_dc = divisive_clustering(X, 2)
y_dc
```

```python
import numpy as np
import matplotlib.pyplot as plt

n_clusters = len(np.unique(y_dc))

centroids = np.zeros((n_clusters, X.shape[1]))
for i in range(n_clusters):
    centroids[i] = X[y_dc == i].mean(axis=0)

plt.figure(figsize=(10, 6))
plt.scatter(X[:, 0], X[:, 1], c=y_dc, cmap='viridis', s=50, label='Data Points')
plt.scatter(centroids[:, 0], centroids[:, 1], s=200, c='red', marker='X', label='Centroids')
plt.colorbar(label='Cluster Label')
plt.title('Scatter Plot of Clusters with Centroids')
plt.xlabel('Culmen length')
plt.ylabel('Culmen depth')
plt.legend()
plt.show()
```
