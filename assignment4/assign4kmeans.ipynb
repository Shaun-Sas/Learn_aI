```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import pandas as pd

df = pd.read_csv('data.csv')
print(df.head(10))

# Remove rows with NaN values
df = df.dropna()
print(df.head(10))
```

```python
df.info()
```

```python
inputs = df.iloc[:, 0:2].values
inputs.T
```

```python
plt.figure(figsize = (14, 10))
plt.scatter(inputs[:, 0], inputs[:, 1])
```

```python
from sklearn.impute import SimpleImputer
impute = SimpleImputer()
X = impute.fit_transform(inputs)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = sc.fit_transform(X)
X.T
```

```python
plt.figure()
plt.scatter(X[:, 0], X[:, 1])
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters = 3)
kmeans.fit(X)
```

```python
y_predict = kmeans.predict(X)
y_predict
```

```python
kmeans.cluster_centers_
```

```python
centers = sc.inverse_transform(kmeans.cluster_centers_)
centers
```

```python
plt.figure(figsize = (14, 10))
plt.scatter(inputs[y_predict == 0, 0], inputs[y_predict == 0, 1], s = 100, c = 'blue', label = 'Cluster 1') #for first cluster
plt.scatter(inputs[y_predict == 1, 0], inputs[y_predict == 1, 1], s = 100, c = 'green', label = 'Cluster 2') #for second cluster
plt.scatter(inputs[y_predict == 2, 0], inputs[y_predict == 2, 1], s = 100, c = 'red', label = 'Cluster 3') #for third cluster
plt.scatter(centers[:, 0], centers[:, 1], s = 250, c = 'k', label = 'Centroid')
plt.title('Clusters of penguins')
plt.xlabel('Culmen length')
plt.ylabel('Culmen depth')
plt.legend()
plt.show()
```

```python
wcss_list = []  # Initializing the list for the values of WCSS

# Using for loop for iterations from 1 to 10.
for i in range(1, 11):
  kmeans = KMeans(n_clusters = i)
  kmeans.fit(X)
  wcss_list.append(kmeans.inertia_)
  print(f'current inertia value with k value {i} is {wcss_list[i-1]}')
  # print(f'centroid point is: {kmeans.cluster_centers_}')
plt.plot(range(1, 11), wcss_list)
plt.title('The Elobw Method Graph')
plt.xlabel('Number of clusters(k)')
plt.ylabel('wcss_list')
```

```python
from sklearn.metrics import silhouette_score

silhoutte_list = []

for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, random_state=47)
    labels = kmeans.fit_predict(X)
    silhouette_scores = silhouette_score(X, labels)
    average_silhouette_score = np.mean(silhouette_scores)
    silhoutte_list.append(average_silhouette_score)
    print("k = {} | silhouette score = {}".format(k, average_silhouette_score))
plt.plot(range(2, 11), silhoutte_list)
plt.title('The Silhoutte Method Graph')
plt.xlabel('Number of clusters(k)')
plt.ylabel('silhoutte_list')
```
